```python
dataset_ = "~/Box/Huang_lab/Huang_lab_data/PancanIOTmb_Samstein_NatGene2019/"
filename = "data_mutations_mskcc.txt"
dataset_path = pd.read_csv(dataset_ + filename, sep = '\t', dtype = 'object')

tmb = "data_clinical_patient.txt"
dataset_tmb = pd.read_csv(dataset_ + tmb, sep = '\t', dtype = 'object')
print(dataset_tmb.columns)
```

```python
dataset_path.tail
```

```python
dataset_path.columns
```

```python
genes_col = dataset_path['Hugo_Symbol']
```

```python
genes = np.unique(genes_col)
```

```python
samp_col = dataset_path['Tumor_Sample_Barcode']
```

```python
samp = np.unique(samp_col)
```

```python
samp_col[0][0:9]
```

```python
data_mat = np.zeros((len(samp), len(genes)))
data_mat.shape
```

```python
for i in range(0, len(samp_col)):
    
    row = np.where(samp == samp_col[i])[0][0]
    
    col = np.where(genes == genes_col[i])[0][0]
    #print(row, col)
    data_mat[row][col] = 1
```

```python
sum_gene = np.sum(data_mat, axis=0)
sns.distplot(sum_gene, bins=20, kde=False, rug=True);
```

```python
print(sum_gene)
print(len(sum_gene))
print(np.where(sum_gene>100))
print(genes[np.where(sum_gene>100)])
sns.barplot(x=list(range(0, len(sum_gene))),y=sum_gene)
```

```python
sns.set(rc={'figure.figsize':(11.7,85)})

ax = sns.barplot(y=genes, x = sum_gene)
plt.savefig('./out/mskcc/genes_mutations.pdf')
```
